Finding the first “Spotlight
deal” on the website. Further, printing the old and reduced price of the
same. (For example, the first spotlight deal is for the product let’s say, Acer
Chromebook. The original price was 249.99 dollars, and the new reduced deal price is 129.99 dollars. 

#### Import Packages

```python
from bs4 import BeautifulSoup
import requests
import lxml
import re
import imgkit
import IPython
```

#### Capture Ebay deals page

```python
try:
    ## Url to load from
    url= "https://www.ebay.com/deals/"

    ## Sotring the get request result in a variable
    page = requests.get(url)
except:
    print("Problem with the connection...")
```

#### Converting the page request into beautiful soup element

```python
## Converting the request page in the text format
soup = BeautifulSoup(page.text, 'lxml')

## Get the contents of class having spotlight deal
list_of_contents = soup.findAll("div", {"class": "ebayui-dne-summary-card card ebayui-dne-item-featured-card--topDeals ebayui-dne-featured-with-carousel"})
```

```python
## String the text of the deal into a string
spotlight_deal = BeautifulSoup(str(list_of_contents), 'lxml')
```

#### Capturing the First Spotlight Deal - Name, Old and New price

```python
## Capturing Product Name
try:
    Product_Name = spotlight_deal.findAll("span", {"class": "ebayui-ellipsis-3"},)
    print("Product Name : " + Product_Name[0].text)
except:
    print("Product Name not available")

## Dicounted Price / Reduced Price
try:
    discounted_price = spotlight_deal.findAll("span", {"class": "first"})
    print("Discounted Price : " + discounted_price[0].text)
except:
    print("Discounted Price is not available")

## Old Price
try:
    original_price = spotlight_deal.findAll("span", {"class": "itemtile-price-strikethrough"})
    print("Original Price : " + original_price[0].text)
except:
    print("Original Price is not avaiable")
```
